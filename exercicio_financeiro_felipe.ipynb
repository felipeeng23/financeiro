```python
'''1 - Filtre os clientes de São Paulo e os exiba em um df pandas
'''
import pandas as pd
from pymongo import MongoClient

# Conectando ao MongoDB Atlas (substitua pela sua URI)
client=MongoClient('mongodb+srv://felipehoobastank:Sapoti2025@cluster0.wuzvndr.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0');

# Acessando o banco e a coleção
db = client["financeiro"]  # Exemplo: db = client["meu_banco"]
colecao_clientes = db["cliente"]

# Carregando os dados da coleção cliente
clientes_data = list(colecao_clientes.find())

# Convertendo para DataFrame
df_clientes = pd.DataFrame(clientes_data)

# Normalizando e filtrando clientes de São Paulo
if 'Municipio' in df_clientes.columns:
    df_clientes['Municipio'] = df_clientes['Municipio'].str.upper().str.strip()
    df_clientes_sp = df_clientes[df_clientes['Municipio'] == 'SAO PAULO']
else:
    df_clientes_sp = pd.DataFrame()  # retorna vazio se a coluna não existir

# Exibindo o resultado
print(df_clientes_sp)


```

```python
'''Faça uma consulta que retorne todos os fornecedores do tipo pessoa física
exiba em um df pandas
'''
import pandas as pd
from pymongo import MongoClient

# Conectando ao MongoDB Atlas
client=MongoClient('mongodb+srv://felipehoobastank:Sapoti2025@cluster0.wuzvndr.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0');

colecao_fornecedores = db["fornecedor"]

# Buscando todos os fornecedores do tipo Pessoa Física (case-insensitive)
fornecedores_data = list(colecao_fornecedores.find({
    'Tipo Pessoa': {'$regex': '^pessoa física$', '$options': 'i'}
}))

# Criando o DataFrame
df_fornecedores_pf = pd.DataFrame(fornecedores_data)

# Exibindo todas as linhas de forma amigável
pd.set_option('display.max_rows', None)
display(df_fornecedores_pf)

```
